# Checking dataset is normally distributed or not concerning that which values are reliable to fill the null values 

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt



# Create a list of numeric columns (excluding 'Time Period')
numeric_columns = [col for col in df.columns if col != 'Time Period' and np.issubdtype(df[col].dtype, np.number)]

# Perform normality tests and plot histograms and QQ plots
for column in numeric_columns:
    data = df[column].dropna()
    # Shapiro-Wilk test
    shapiro_test = stats.shapiro(data)
    # Anderson-Darling test
    anderson_test = stats.anderson(data)
    # D'Agostino and Pearson's test
    dagostino_test = stats.normaltest(data)
    
    print(f"Column: {column}")
    print(f"Shapiro-Wilk Test - p-value: {shapiro_test.pvalue}")
    print(f"Anderson-Darling Test - Statistic: {anderson_test.statistic}, Critical Values: {anderson_test.critical_values}")
    print(f"D'Agostino and Pearson's Test - p-value: {dagostino_test.pvalue}")
    
    # Histogram
    plt.figure(figsize=(10, 4))
    plt.subplot(1, 2, 1)
    plt.hist(data, bins='auto', density=True)
    plt.title('Histogram')
    
    # QQ Plot
    plt.subplot(1, 2, 2)
    stats.probplot(data, dist='norm', plot=plt)
    plt.title('QQ Plot')
    
    plt.show()
```

As we can see from the data statistical tests and qq plot and the histogram the dataset is not normally distributed therefore it wolud be accepted that that null values of the dataset can be filled using the median of the values

# Using different scalling methods and trying to understand which method is best for thius dataset using cross validation method 

```python
from sklearn.model_selection import cross_val_score
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import RobustScaler, MinMaxScaler, QuantileTransformer, PowerTransformer
from sklearn.linear_model import LinearRegression
import warnings 

# Features and target variable
X = df1[['max_temp', 'mean_temp', 'precipitation', 'cloud_cover', 'sunshine']]
y = df1['convenient stores '] 

# Create a dictionary of scalers
scalers = {
    'RobustScaler': RobustScaler(),
    'MinMaxScaler': MinMaxScaler(),
    'QuantileTransformer': QuantileTransformer(),
    'PowerTransformer': PowerTransformer(method='yeo-johnson')
}

results = {}

# Perform cross-validation for each scaler
for scaler_name, scaler in scalers.items():
    pipeline = Pipeline([
        ('scaler', scaler),
        ('model', LinearRegression())  # You can replace this with your desired model
    ])
    
    scores = cross_val_score(pipeline, X, y, cv=5, scoring='neg_mean_squared_error')
    results[scaler_name] = scores.mean()

# Print the results
for scaler_name, score in results.items():
    print(f'{scaler_name}: Mean MSE = {score}')
```

Among the provided scaling methods, the least mean negative mean squared error (MSE) is for the PowerTransformer with a value of approximately -27,878,350,146.67. This indicates that the PowerTransformer performed the best among the tested scaling methods based on the cross-validation results.These MSE values are negative because scikit-learn's cross_val_score returns negated MSE values.

# Linear regression analysis to estimate the sensitivity of apparel sales to weather variables.

```python
import numpy as np
import statsmodels.api as sm
import warnings

# Load and prepare data (assuming 'df1' is your DataFrame)
# Ensure that the columns are correctly named in your dataset
df = df1[['Time Period', 'sunshine', 'precipitation', 'max_temp','mean_temp', 'convenient stores ']]

# Extract year and month from 'Time Period' column
df['Year'] = df['Time Period'].str[:4].astype(int)
df['Month'] = df['Time Period'].str[4:].astype(int)

# Loop through each unique month
months = np.arange(1, 13)  # Months from 1 to 12

results = []

for month in months:
    subset = df[df['Month'] == month]
    
    # Split data into features (X) and target variable (y)
    X = subset[['sunshine', 'precipitation', 'max_temp', 'mean_temp']]
    X = sm.add_constant(X)  # Add a constant term for intercept
    y = subset['convenient stores ']
    
    # Perform linear regression
    model = sm.OLS(y, X).fit()
    
    # Store results
    result = {
        'Month': month,
        't_value_sunshine': model.tvalues['sunshine'],
        'p_value_sunshine': model.pvalues['sunshine'],
        't_value_precip': model.tvalues['precipitation'],
        'p_value_precip': model.pvalues['precipitation'],
        't_value_max_temp': model.tvalues['max_temp'],
        'p_value_max_temp': model.pvalues['max_temp'],
        't_value_mean_temp': model.tvalues['mean_temp'],
        'p_value_mean_temp': model.pvalues['mean_temp'],
        'r_squared': model.rsquared,
        'coefficient_sunshine': model.params['sunshine'],
        'coefficient_precip': model.params['precipitation'],
        'coefficient_max_temp': model.params['max_temp'],
        'coefficient_mean_temp': model.params['mean_temp'],
        'f_statistic': model.fvalue,
        'f_p_value': model.f_pvalue
    }
    
    results.append(result)

# Create a DataFrame from the results
results_df = pd.DataFrame(results)

# Print or analyze the results_df as needed
print(results_df)
```

# KNN

```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error

# Assuming 'df1' is already loaded and contains the necessary columns
target_column = 'convenient stores '
features_columns = ['cloud_cover', 'max_temp', 'mean_temp', 'precipitation', 'sunshine']

# Extract features (X) and target variable (Y) from the DataFrame
try:
    X = df1_transformed[features_columns].values
    Y = df1_transformed[target_column].values
except KeyError as e:
    print(f"Column not found: {e}")
    exit(1)  # Exit the script if a column is missing

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.3, random_state=42)

# Initialize the k-NN model
k = 5
knn_model = KNeighborsRegressor(n_neighbors=k)

# Train the k-NN model
knn_model.fit(X_train, y_train)

# Make predictions on the test set
y_pred = knn_model.predict(X_test)

# Evaluate the model's performance
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)

# Print the performance metrics
print(f"Mean Squared Error (MSE): {mse:.2f}")
print(f"R-squared (R2) Score: {r2:.2f}")
print(f"Mean Absolute Error (MAE): {mae:.2f}")
```

# Checking optimal value of k (which point will work better)

```python
from sklearn.model_selection import train_test_split, cross_val_score


# Extract features (X) and target variable (Y) from the DataFrame
X = df1_transformed[features_columns].values
Y = df1_transformed[target_column].values

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.3, random_state=42)

# Search for the optimal k value using cross-validation
best_score = -np.inf
best_k = None
for k in range(1, 21):  # You can adjust the range of k values to test
    knn_model = KNeighborsRegressor(n_neighbors=k)
    scores = cross_val_score(knn_model, X_train, y_train, cv=5, scoring='neg_mean_squared_error')  # 5-fold cross-validation
    avg_score = np.mean(scores)
    if avg_score > best_score:
        best_score = avg_score
        best_k = k

# Train the KNN model with the optimal k value
best_knn_model = KNeighborsRegressor(n_neighbors=best_k)
best_knn_model.fit(X_train, y_train)

# Make predictions on the test set
y_pred = best_knn_model.predict(X_test)

# Evaluate the model's performance
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

# Print the optimal k value and performance metrics
print(f"Optimal k: {best_k}")
print(f"Mean Squared Error (MSE): {mse:.2f}")
print(f"R-squared (R2) Score: {r2:.2f}")
```
