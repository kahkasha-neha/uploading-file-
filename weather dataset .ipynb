```python
import pandas as pd 
df= pd.read_csv('uk_weather.csv')
df.head()
```

```python
# Calculate the mean of each column (excluding null values)
column_means = df.mean()

# Fill null values in all columns with their respective column means
df1 = df.fillna(column_means)

# Display the DataFrame with null values filled with the column means
df1.head()
```

```python
import pandas as pd

# Assuming you have loaded the dataset into a DataFrame called 'df'
# Convert the 'time' column to datetime type
df['date'] = pd.to_datetime(df['date'], format='%Y%m%d')

# Create a new column 'year_month' to store the year and month information
df['Time Period'] = df['date'].dt.strftime('%Y%m')

# Drop the original 'time' column if you don't need it anymore
df.drop(columns=['date'], inplace=True)

# Group by 'year_month' and calculate the mean for each group
df_monthly_avg = df.groupby('Time Period').mean()

# Reset the index to have 'year_month' as a regular column
df_monthly_avg.reset_index(inplace=True)

# Display the DataFrame with average values for each month
df_monthly_avg.head()


```

```python
df_monthly_avg.tail()
```

```python
import pickle

# Assuming you have the 'df_avg' DataFrame containing the averaged data
# Pickle the DataFrame and save it to a file
with open('df_monthly_avg.pkl', 'wb') as f:
    pickle.dump(df_monthly_avg, f)
```
