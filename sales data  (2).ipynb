# changing the dtype of the column into  

```python

# Function to check if a string can be converted to a float
def is_float(x):
    try:
        float(x)
        return True
    except ValueError:
        return False

# Loop through each column in the DataFrame (except 'Time Period')
for column in df.columns:
    if column != 'Time Period':
        # Check if the column is of object data type and if all values can be converted to float
        if df[column].dtype == 'object' and df[column].apply(is_float).all():
            # Convert the column to float
            df[column] = df[column].astype(float)

# Now, the numeric columns (except 'Time Period') in the DataFrame should be of float data type
```

# Checking dataset is normally distributed or not concerning that which values are reliable to fill the null values 

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt



# Create a list of numeric columns (excluding 'Time Period')
numeric_columns = [col for col in df.columns if col != 'Time Period' and np.issubdtype(df[col].dtype, np.number)]

# Perform normality tests and plot histograms and QQ plots
for column in numeric_columns:
    data = df[column].dropna()
    # Shapiro-Wilk test
    shapiro_test = stats.shapiro(data)
    # Anderson-Darling test
    anderson_test = stats.anderson(data)
    # D'Agostino and Pearson's test
    dagostino_test = stats.normaltest(data)
    
    print(f"Column: {column}")
    print(f"Shapiro-Wilk Test - p-value: {shapiro_test.pvalue}")
    print(f"Anderson-Darling Test - Statistic: {anderson_test.statistic}, Critical Values: {anderson_test.critical_values}")
    print(f"D'Agostino and Pearson's Test - p-value: {dagostino_test.pvalue}")
    
    # Histogram
    plt.figure(figsize=(10, 4))
    plt.subplot(1, 2, 1)
    plt.hist(data, bins='auto', density=True)
    plt.title('Histogram')
    
    # QQ Plot
    plt.subplot(1, 2, 2)
    stats.probplot(data, dist='norm', plot=plt)
    plt.title('QQ Plot')
    
    plt.show()
```

As we can see from the data statistical tests and qq plot and the histogram the dataset is not normally distributed therefore it wolud be accepted that that null values of the dataset can be filled using the median of the values  

# Filling the null values of the dataset 

```python
# Calculate the mean of each column (excluding null values)
column_med = df.median()

# Fill null values in all columns with their respective column means
df1 = df.fillna(column_med)

# Display the DataFrame with null values filled with the column means
df1.head()
```

# Diving the time period column into year and month column to justify which month and which year has maximum sales

```python
# Assuming 'Time Period' is in the format "1994JAN" to "1994DEC"
df['year'] = df['Time Period'].str[:4].astype(int)
df['month'] = df['Time Period'].str[4:]
```

# filter the year 1994 and calculate the sales for each month

```python
sales_1994 = df[df['year'] == 1994].groupby('month')['convenient stores '].sum()
```

# plot the dataset as bar chart


```python
import matplotlib.pyplot as plt
# Set the style for the plot (optional)
plt.style.use('ggplot')

# Create the bar chart
plt.figure(figsize=(10, 6))
plt.bar(sales_1994.index, sales_1994.values)
plt.xlabel('Month')
plt.ylabel('Sales')
plt.title('convenient stores ')
plt.xticks(rotation=45)

# Display the plot
plt.tight_layout()
plt.show()
```

# making a dashboard to know the sales of convinient stores of form 1994-2023

```python
import dash
from dash import dcc, html
import plotly.express as px
```

```python
# Extract the year and month from 'Time Period' column
df['year'] = df['Time Period'].str[:4].astype(int)
df['month'] = df['Time Period'].str[4:]
```

```python
#initialize dash app 
app = dash.Dash(__name__)
```

```python
app.layout = html.Div([
    html.H1("convenient stores  "),
    dcc.Graph(id='sales-chart')
])
```

```python
@app.callback(
    dash.dependencies.Output('sales-chart', 'figure'),
    [dash.dependencies.Input('year-dropdown', 'value')]
)
def update_chart(selected_year):
    filtered_df = df[df['year'] == selected_year]
    fig = px.bar(filtered_df, x='month', y='convenient stores ', title=f"Sales in {selected_year}",
                 labels={'convenient stores ': 'Sales', 'month': 'Month'})
    return fig
```

```python
year_options = [{'label': str(year), 'value': year} for year in df['year'].unique()]

app.layout = html.Div([
    html.H1("convenient stores "),
    dcc.Dropdown(
        id='year-dropdown',
        options=year_options,
        value=df['year'].min(),  # Default value
        style={'width': '50%'}
    ),
    dcc.Graph(id='sales-chart')
])
```

```python
if __name__ == '__main__':
    app.run_server(debug=True, port=8051)
```
